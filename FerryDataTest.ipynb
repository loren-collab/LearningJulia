```julia
# Packages required
using IJulia
using DataFrames
using CSV
using Plots
```

```julia
# The file with the interesting data is loaded
data = CSV.read("C:\\Users\\Lorenzo\\Documents\\MyProjects\\Complete.csv", DataFrame);
n = size(data, 1)
time = [15 * i for i in 0:n-1]; 
```

```julia
data_reduced = data[:, 1:5]
rename!(data_reduced, 1 => "Timestamp", 2 => "Thuster 1", 3 => "Thuster 2", 4 => "Thuster 3", 5 => "Thuster 4")
```

```julia
data_reduced[!, :Seconds] = time;
data_reduced
```

```julia
plot(time/60, data_reduced[:,2], xlabel="Time [min]", ylabel="Power [kW]", label=names(data_reduced)[2], title="Thruster Power", dpi=1000)
plot!(time/60, data_reduced[:, 3], label=names(data_reduced)[3])
plot!(time/60, data_reduced[:, 4], label=names(data_reduced)[4])
plot!(time/60, data_reduced[:, 5], label=names(data_reduced)[5])
```
